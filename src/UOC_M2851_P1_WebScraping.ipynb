```python
from bs4 import BeautifulSoup
import requests
import sys, os
import csv
import random, time
```

```python
# Auxiliary functions
#
def tiempos_control_espera(tipo_control):
    #Performs scheduled stops with random times in ranges
    if tipo_control=="pequeño":
        tiempos=[10,11,12,20,21,22,30,31]
    elif tipo_control=="mediano":
        tiempos=[60,70,80,90,100,200,300,400]
    else:
        tiempos=[1000,2000,3000,4000,5000,6000,7000,8000]
    tiempo_aleatorio=tiempos[random.randrange(7)]
    if tipo_control=="grande":
        print('Tiempo de espera {}, para no sobrepasar límites: {} '.format(tipo_control, tiempo_aleatorio))
    time.sleep(tiempo_aleatorio)
#
def inicializar_variables():
    # initializes  variables.
    #
    pais='España'
    competicion='La Liga'
    categoria='1ª división'
    #
    temporadaInicial=2012
    temporadaFinal=2018
    #
    urlBase ='https://www.sport.es/es/resultados/'
    #
    filePath='download/soccer/'
    if not os.path.exists(filePath):
        os.makedirs(filePath)
    processedFile=filePath+"estadisticasFutbol.csv"
    #
    headers = {
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,\*/*;q=0.8",
        "Accept-Encoding": "gzip, deflate, sdch, br",
        "Accept-Language": "en-US,en;q=0.8",
        "Cache-Control": "no-cache",
        "dnt": "1",
        "Pragma": "no-cache",
        "Upgrade-Insecure-Requests": "1",
        "User-Agent": "Mozilla/5.0 (Windows NT 6.3; Win64; x64; rv:63.0) Gecko/20100101 Firefox/63.0"
    }
    return pais,competicion,categoria,temporadaInicial,temporadaFinal,urlBase,processedFile,headers
#
```

```python
# Main program
# Get league soccer results since 2012
def getSoccerResults():
    #
    pais,competicion,categoria,temporadaInicial,temporadaFinal,urlBase,processedFile,headers=inicializar_variables()
    #
    with open(processedFile, "w") as myfile:
        #headers of the csv file
        writer = csv.writer(myfile)
        fieldnames = ['País', 'Competición', 'Categoría', 'Temporada','Jornada','Equipo 1','Resultado','Equipo 2']
        writer = csv.DictWriter(myfile, fieldnames=fieldnames)
        writer.writeheader()
        #
        for i in range(temporadaInicial,temporadaFinal):
            #the urls are built...
            temporadaActual =str(i)+'-'+str(i+1)
            urlAux=urlBase+str(temporadaActual)
            print (temporadaActual)  
            for j in range(1,39):
                urlActual= urlAux+'/la-liga/jornada-'+ str(j)+'/'
                print('Jornada' + str(j)+': '+urlActual)  
                #
                #csv file content
                try:
                    r = requests.get(urlActual,headers=headers, timeout=10)
                    if r.status_code != 200:
                        print('\t {}, no procesada. '.format(urlActual))
                    else:
                        soup = BeautifulSoup(r.content)
                        ul = soup.find('table', attrs={'class':'table'})
                        for Partido in ul.find_all('tr'):
                            td1 = Partido.find(attrs={'class':'td1'})
                            equipo1 = td1.text 
                            #
                            td2 = Partido.find(attrs={'class':'td2'})
                            resultado = td2.text 
                            #
                            td3 = Partido.find(attrs={'class':'td3'})
                            equipo2 = td3.text 
                            #
                            #we add a row with content to the csv file    
                            writer.writerow({'País': pais, 'Competición': competicion,'Categoría': categoria, 'Temporada': temporadaActual,'Jornada':str(j),'Equipo 1': equipo1, 'Resultado': resultado,'Equipo 2': equipo2})
                        #
                #
                except  requests.exceptions.Timeout as errTimeOut: 
                    print('/tfailed to download url:{} por {}:{} '.format(urlActual,errTimeOut,errTimeOut.args))
                except  requests.exceptions.RequestException as errRequestException: 
                    print('/tfailed to download url:{} por {}:{} '.format(urlActual,errRequestException,errRequestException.args))
                #small break
                tiempos_control_espera("pequeño")
            #medium break
            tiempos_control_espera("mediano")           
        myfile.close() 
```

```python
getSoccerResults()
print('oo')
```
